```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
# creating the dataset
X,Y= make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=15)
```

```python
pd.DataFrame(X)
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y, test_size=0.30, random_state=42)
```

```python
# Model Training
from sklearn.linear_model import LogisticRegression
logistic= LogisticRegression()

logistic.fit(X_train,Y_train)
```

```python
Y_pred= logistic.predict(X_test)
print(Y_pred)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
score= accuracy_score(Y_test,Y_pred)
print(f"Accuracy Score: {score}")
cm= confusion_matrix(Y_test,Y_pred)
print(f"Confusion Matrix:\n{cm}")
report= classification_report(Y_test,Y_pred)
print(f"Classification Report:\n{report}")
```

### Hyperparameter Tunning and Cross-Validation

```python
model = LogisticRegression()
penalty = ['l1', 'l2', 'elasticnet']
c_values = [0.01, 0.1, 1, 10, 100]
solver = ['liblinear', 'saga', 'lbfgs', 'newton-cg', 'sag']
```

```python
params = dict( penalty= penalty, C= c_values, solver= solver)
```

```python
## Grid Search CV
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
grid_search = GridSearchCV(estimator=model, param_grid=params, cv=cv, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, Y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
Y_pred= grid_search.predict(X_test)
```

```python
score= accuracy_score(Y_test,Y_pred)
print(f"Accuracy Score: {score}")
cm= confusion_matrix(Y_test,Y_pred)
print(f"Confusion Matrix:\n{cm}")
report= classification_report(Y_test,Y_pred)
print(f"Classification Report:\n{report}")
```

### Randomizsed Search Cross-Validation

```python
from sklearn.model_selection import RandomizedSearchCV
model = LogisticRegression()
random_search = RandomizedSearchCV(estimator=model, param_distributions=params, cv=5, scoring='accuracy', n_jobs=-1)
random_search.fit(X_train, Y_train)
```

```python
random_search.best_score_
```

```python
random_search.best_params_
```

```python
Y_pred= random_search.predict(X_test)
```

```python
score= accuracy_score(Y_test,Y_pred)
print(f"Accuracy Score: {score}")
cm= confusion_matrix(Y_test,Y_pred)
print(f"Confusion Matrix:\n{cm}")
report= classification_report(Y_test,Y_pred)
print(f"Classification Report:\n{report}")
```
